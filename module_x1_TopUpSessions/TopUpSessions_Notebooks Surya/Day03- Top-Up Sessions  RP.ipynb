# Random Forest
1. ```Bagging```
    - Decision Trees are in parallel, collective decision making with rule of majority
    - works with multiple trees in a forest
    - P(Decision Tree giving correct result), ```P(True)``` = 0.6
    - No of trees, ```n``` = 5
    - No of Trees say True, ```r``` = 3
    - Probability of ```r``` Trees saying True out of ```n``` trees = ```nCr * P(True)^r * (1-P(True))^(n-r)``` [```using Binomial Distribution```]
    - Correct Value = True
    - If three of more forest say that it is True
        - ```P(One forest will say True)``` = ```P(Three or more trees say True)```
            - ```P('r' trees say True out of 'n' Trees)``` = ```nCr * P(True)^r * (1-P(True))^(n-r)```
            - ```P(Three trees say True)``` = ```5C3 * 0.6^3 * (1-0.6)^(5-3)```
                - = ```5C3 * 0.6^3 * 0.4^2```
                - = ```0.3456```
            - ```P(Four trees say True)``` = ```5C4 * 0.6^4 * (1-0.6)^(5-4)```
                - = ```5C4 * 0.6^4 * 0.4^1```
                - = ```0.2592```
            - ```P(Five trees say True)``` = ```5C5 * 0.6^5 * (1-0.6)^(5-5)```
                - = ```5C5 * 0.6^5 * 0.4^0```
                - = ```0.07776```
            - ```P(One forest will say True)``` = ```P(Three trees say True)``` + ```P(Four trees say True)``` + ```P(Five trees say True)```
                - = ```0.3456``` + ```0.2592``` + ```0.07776```
                - = ```0.68256```
- Bootstrapping
    - another form of Bagging
    - works with multiple forests

    - P(Decision Tree giving correct result), ```P(True)``` = 0.6
    - No of trees, ```n``` = 5
    - No of Trees say True, ```r``` = 3
    - Probability of ```r``` Forests saying True out of ```n``` Forests = ```nCr * P(True)^r * (1-P(True))^(n-r)``` [```using Binomial Distribution```]
    - Correct Value = True
    - If three of more forest say that it is True
        - ```P(One forest will say True)``` = ```P(Three or more trees say True)```
            - ```P('r' trees say True out of 'n' Trees)``` = ```nCr * P(True)^r * (1-P(True))^(n-r)```
            - ```P(Three Forests say True)``` = ```5C3 * 0.6^3 * (1-0.6)^(5-3)```
                - = ```5C3 * 0.6^3 * 0.4^2```
                - = ```0.320439```
            - ```P(Four Forests say True)``` = ```5C4 * 0.6^4 * (1-0.6)^(5-4)```
                - = ```5C4 * 0.6^4 * 0.4^1```
                - = ```0.148151```
            - ```P(Five Forests say True)``` = ```5C5 * 0.6^5 * (1-0.6)^(5-5)```
                - = ```5C5 * 0.6^5 * 0.4^0```
                - = ```0.07776```
            - ```P(One set of Forests will say True)``` = ```P(Three Forests say True)``` + ```P(Four Forests say True)``` + ```P(Five Forests say True)```
                - = ```0.320439``` + ```0.344505``` + ```0.148151```
                - = ```0.813095```
        - Note that, in aggregation probability of correct prediction has increased
2. ```Boosting```
    - Decision Trees are put in series and keep on boosting
    - helps weak learners to become strong learners by making weak learners learn again in the series, so last learner is a strong learner
    - weights (```α1```, ```α2```, ```α3```, ```...```) are assigned to each learner, and weights are adjusted after each learner to reduce error
    - keeps on aggregating repetitively
    
    - Two types of boosting
        1. ```Gradient Boosting / Descent```
            - Hyper Parameters
                - Alpha, ```α``` : Learning Rate / Speed of convergance /  [between ```0 and 1```]
            - Steps
            1. Create a base model, Average of all values
            2. Calculate the residuals based on predicted value from base model
            3. Create Decision Tree (DT1) with the inputs as Experience, Education and output as Residuals(D)
            4. Generate Predictions for DT1
            5. Calculate resuduals (resid2) based on predicted values of DT1
            6. ```New Prediction``` = ```Previous prediction``` + ```Prediction from tree based on residuals(NewPred1A)```
            7. To avoid overfitting the previous equation of Row#1 is changed to
                - ```New Prediction``` = ```Previous Prediction``` + ```Learning Rate, α``` * ```Prediction from tree based on Residuals```
                - Learning Rate, ```α``` will always be between ```0 and 1```, preferable around ```0.1 to 0.15```
                - Low value of Learning rate makes take small steps, so it takes more time
                - High value of Learning rate make take big steps, so it takes less time
        2. ```XG Boosting``` / ```eXtreme Gradient Boosting```
            - Hyperparameters / Terminology
                - Lambda, ```λ``` : L2 Regularization Parameter
                    - analogous to Ridge regression
                    - can control Gain to reduce the Similarity Score to ultimately reduce the Gain below Pruning threshold(```γ```)
                    - decides how many split can be made
                    - if ```λ``` = 0, outliers have more effect, so to reduce effect of outliers or even errors, we introduce Regularization parameter ```λ```
                    - with introduction of regularization parameter ```λ```, we can reduce the impact of outliers
                - Eta, ```η``` : Learning Rate / Speed of convergance
                    - between ```0 and 1```
                    - Analogous to the learning rate in GBM
                - Gamma, ```γ``` : Threshold of Pruning
                - ```Similarity Score 1``` = ```Square(sum(residuals))/(No of values)```
                    - = ```(∑ Residuals)^2``` / ```No of values```
                - ```Similarity Score 2``` = ```Square(sum(residuals))/(No of values + Regularization Parameter) ```
                    - = ```(∑ Residuals)^2``` / ```(No of values + λ)```
            - Steps
                1. Create a base model to make the first prediction
                2. we consider the first prediction
                3. we decide the criteria as Exp>=5
