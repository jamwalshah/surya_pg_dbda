# Optimizers for Artificial Neural Networks
1. ```Gradient Descent``` (```GD```)
    - ```W(new)``` = ```W(old)``` - ```η``` * ```ðLoss/ðW```
    - ```Epoch```
        - no of times records are passed
        - huge amount of resources (RAM) are required
        - convergence is quick so less time complexity
2. ```Stochastic Gradient Descent``` (```SGD```)
    - single record is passed in one epoch
    - Convergence is slow, time complexity will be high
3. Mini-Batch Stochastic Gradient Descent (```Mini-Batch SGD```)
    - pass multiple records (say 1000 records) as batches in one epoch
    - number of iterations would be lesser as compared to SGD and convergence would be faster
4. ```Stochastic Gradient Descent with momentum```
    - use exponential smoothing to reduce the zig-zag movement for reaching the minima
5. ```Adaptive Gradient Descent``` (```AdaGrad```)
    - action taken on Learning Rate```η```
    - ```η'``` = ```η``` / ```√(α t + ϵ)```
        - where ```α t``` = ```i=1→t∑```(```ðLoss```/```ðW t```)```^2```
    - ```η'``` is very small
- Chain Rule of differentiation
    - ```ðL/ðW1``` = ```ðLoss/ðw2``` * ```ðw2/ðw1```
    - ```Vanishing Gradient Problem```
        - when weights are update in multiple layers, all the derivatives get multiplied using chain diff rule, and all diff are between 0  and 1, so new diff  becomes very low
    - ```Exploding Gradient Problem```
        -
        - for sigmod function
            - only function which gives values between 0 and 1
            - derivative of sigmoid function will always be between 0 and 1
        - for ReLu Function
            - derivative of Relu is 0 if x=0
            - max(0, 1), either 0 if value is -ve or 1 if value is +ve
            - either 0 or the value itself
            - Problem with Relu
                - in case of back propagation, if derivative of Relu is 0, that neuron becomes dead, and it can used to advantage to avoid overfitting

## load dataset

````python
df = pd.read_csv("Churn_Modelling.csv")
````

````python
df.head()
````

## drop unwanted commands

````python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
````

````python
df.head()
````

## X & Y

````python
x = df.drop('Exited', axis=1)
````

````python
y = df['Exited']
````

## Create dummies for categorical columns

````python
geo_dummy = pd.get_dummies(x['Geography'], drop_first=True)
````

````python
gender_dummy = pd.get_dummies(x['Gender'], drop_first=True)
````

````python
x = x.drop(['Geography', 'Gender'], axis=1)
````

````python
x = pd.concat([x, geo_dummy, gender_dummy], axis=1)
````

````python
x.head()
````

## Splitting

````python
from sklearn.model_selection import train_test_split
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
````

## Scaling

````python
from sklearn.preprocessing import StandardScaler
````

````python
sc = StandardScaler()
````

````python
x_train = sc.fit_transform(x_train)
````

````python
x_test = sc.transform(x_test)
````

````python
from tensorflow.keras.models import Sequential
# helps in forward ans backward propagation
````

````python
classifier = Sequential()
````

````python
from tensorflow.keras.layers import Dense
# helps in creating input, output and the hidden layers
````

````python
from tensorflow.keras.layers import ReLU
# helps in using ReLU function
````

````python
classifier.add(Dense(units=11, activation='relu'))
# defining the input layer
````

````python
classifier.add(Dense(units=7, activation='relu'))
# defining the first hidden layer
````

````python
classifier.add(Dense(units=6, activation='relu'))
# defining the second hidden layer
````

````python
classifier.add(Dense(units=1, activation='sigmoid'))
# defining the output layer
````

````python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

````python
mod1 = classifier.fit(x_train, y_train, batch_size=10, epochs=3)
````

````python
y_pred = classifier.predict(x_test)
````

````python
y_pred = y_pred>0.5
````

````python
from sklearn.metrics import confusion_matrix, accuracy_score
````

````python
confusion_matrix(y_test, y_pred)
````
