Exercise 4: Arrange string characters such that
lowercase letters should come first

Given string contains a combination of the lower and upper case letters. Write
a program to arrange the characters of a string so that all lowercase letters
should come first.

```python
'''
Given:
str1 = PyNaTive

Expected Output:
yaivePNT
'''
s1=input("Enter a string: ")
s2=""
for char in s1:
  if char.islower():
    s2=s2+char
for char in s1:
  if char.isupper():
    s2=s2+char
print(s2)
```

Exercise 5: Count all letters, digits, and special
symbols from a given string

```python
'''
Given:
str1 = "P@#yn26at^&i5ve"

Expected Outcome:
Total counts of chars, digits, and symbols
Chars = 8
Digits = 3
Symbol = 4
'''
s1=input("Enter a string: ")
char, digi, symb=0, 0, 0
for ch in s1:
  if ch.isalpha():
    char+=1
  elif ch.isdigit():
    digi+=1
  else:
    symb+=1
print("Total counts of chars, digits, and symbols\nChars =", char, "\nDigits =", digi, "\nSymbols =", symb)
```

Exercise 6: Create a mixed String using the following
rules

Given two strings, s1 and s2. Write a program to create a new string s3 made
of the first char of s1, then the last char of s2, Next, the second char of s1 and
second last char of s2, and so on. Any leftover chars go at the end of the
result.

```python
'''
Given:
s1 = "Abc"
s2 = "Xyz"

Expected Output:
AzbycX
'''
s1=input("Enter first string: ")
s2=input("Enter second string: ")
s3=""
i, j = 0, len(s2)-1
while i<len(s1) or j>0:
  s3=s3+s1[i]
  i+=1
  s3=s3+s2[j]
  j-=1
print(s3)
```

Exercise 7: String characters balance Test

Write a program to check if two strings are balanced. For example, strings s1
and s2 are balanced if all the characters in the s1 are present in s2. The
character’s position doesn’t matter.

```python
'''
Given:
Case 1:
s1 = "Yn"
s2 = "PYnative"

Expected Output:
True

Case 2:
s1 = "Ynf"
s2 = "PYnative"

Expected Output:
False
'''
s1=input("Enter first string: ")
s2=input("Enter second string: ")
flag=True
for ch in s1:
  if ch not in s2:
    flag=False
print(flag)
```

Exercise 9: Calculate the sum and average of the
digits present in a string

Given a string s1, write a program to return the sum and average of the digits
that appear in the string, ignoring all other characters.

```python
'''
Given:
str1 = "PYnative29@#8496"

Expected Outcome:
Sum is: 38 Average is 6.333333333333333
'''
str1 = "PYnative29@#8496"
sum,cnt=0,0
for ch in str1:
  if ch.isdigit():
    sum=sum+int(ch)
    cnt+=1
print("Sum is:",sum, "Average is", sum/cnt)
```

Exercise 10: Write a program to count occurrences of
all characters within a string

```python
'''
Given:
str1 = "Apple"

Expected Outcome:
{'A': 1, 'p': 2, 'l': 1, 'e': 1}
'''
str1 = "Apple"
chlst=[]
cntd={}
for ch in str1:
  chlst.append(ch)
set1=set(chlst)
for se in sorted(set1):
  cntd.update({se:chlst.count(se)})
print(cntd)
```

Exercise 14: Remove empty strings from a list of
strings

```python
'''
Given:
str_list = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]

Expected Output:
Original list of sting
['Emma', 'Jon', '', 'Kelly', None, 'Eric', '']

After removing empty strings
['Emma', 'Jon', 'Kelly', 'Eric']
'''
str_list = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]
new_lst=[]
for item in str_list:
  if item!=None:
    if len(item)>0:
      new_lst.append(item)
print(new_lst)    
```

Exercise 15: Remove special symbols / punctuation
from a string

```python
'''
Given:
str1 = "/*Jon is @developer & musician"

Expected Output:
"Jon is developer musician"
'''
str1 = "/*Jon is @developer & musician"
str2=""
wlst=str1.split()
for ele in wlst:
  for ch in ele:
    if ch.isalnum()==False:
      str2=str2+""
    else:
      str2=str2+ch
  str2=str2+" "
print(str2)
```

Exercise 17: Find words with both alphabets and
numbers

Write a program to find words with both alphabets and numbers from an
input string.

```python
'''
Given:
str1 = "Emma25 is Data scientist50 and AI Expert"

Expected Output:
Emma25
scientist50
'''
str1 = "Emma25 is Data scientist50 and AI Expert"
lst=str1.split()
for word in lst:
  alpchk, numchk=False, False
  for ch in word:
    if ch.isalnum():
      alpchk=True
      break
  for ch in word:
    if ch.isdigit():
      numchk=True
      break
  if alpchk and numchk:
    print(word)
```

Exercise 18: Replace each special symbol with # in
the following string

```python
'''
Given:
str1 = '/*Jon is @developer & musician!!'

Expected Output:
##Jon is #developer # musician##
'''
str1 = '/*Jon is @developer & musician!!'
str2=""
wlst=str1.split()
for ele in wlst:
  for ch in ele:
    if ch.isalnum()==False:
      str2=str2+"#"
    else:
      str2=str2+ch
  str2=str2+" "
print(str2)
```
