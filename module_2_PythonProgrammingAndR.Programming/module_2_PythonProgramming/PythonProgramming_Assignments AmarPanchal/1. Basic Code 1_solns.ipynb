11. Find simple interest

```python
print("Simple Interest: ", (float(input("Enter Principal: "))*float(input("Enter Rate of Interest per annum: "))*float(input("Enter time unit in years: ")))/100)
```

12. Find compound interest

```python
p=float(input("Enter Principal: "))
r=float(input("Enter rate of interest per annum: "))
t=float(input("Enter time unit in years: "))
print("Compund Interest: ", (p*(1+(r/100))**t)-p)
```
