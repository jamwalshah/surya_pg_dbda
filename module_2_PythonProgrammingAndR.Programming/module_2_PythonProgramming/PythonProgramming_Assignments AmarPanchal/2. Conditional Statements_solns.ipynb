16. Check type of triangle

```python
anga=float(input("Enter angle A of Triangle: "))
angb=float(input("Enter angle B of Triangle: "))
angc=float(input("Enter angle C of Triangle: "))
if (anga+angb+angc==180):
  if anga==60 and angb==60 and angc==60:
    print("Equilateral Triangle")
  elif anga==90 or angb==90 or angc==90 :
    print("Right angled Triangle")
  elif (anga>90 and angb+angc<90) or (angb>90 and anga+angc<90) or (angc>90 and angb+anga<90):
    print("Obtuse angled Triangle")
  else:
    print("Acute angled Triangle")
else:
  print("Not a triangle")
```
