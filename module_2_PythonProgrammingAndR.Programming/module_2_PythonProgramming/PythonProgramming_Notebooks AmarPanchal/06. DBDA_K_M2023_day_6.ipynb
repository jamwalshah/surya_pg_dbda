```python
def fact(no):
    if no<=1:
        return 1
    else:
        return no*fact(no-1)
```

```python
#if no1%no2==0 then no2 is gcd else let no1=no2 and no2 be no1%no2
def gcd(no1,no2):
    if no1%no2==0:
        return no2
    else:
        return gcd(no2,no1%no2)
```

```python
gcd(18,12)
```

```python
def Descending(no1,no2,no3):
    '''Takes 2 inputs and prints maximum of 2'''
    if no1>no2 and no1>no3:
        if no2>no3:
            return no1,no2,no3
        else:
            return no1,no3,no2
    elif no2>no1 and no2>no3:
        if no1>no3:
            return no2,no1,no3
        else:
            return no2,no3,no1
    elif no3>no1 and no3>no2:
        if no1>no2:
            return no3,no1,no2
        else:
            return no3,no2,no1
```

```python
Descending(5,6,6)
```

```python
def fd(amount,roi):
    y=1
    print("at end of a year:")
    yield (amount*(roi/100)*y)
    y+=1
    print("at end of 2nd year:")
    yield (amount*(roi/100)*y)
    y+=1
    print("at end of 3rd year:")
    yield (amount*(roi/100)*y)
```

```python
#global variable accessed by all
x=10
def test():
    print("inside test:x=",x)

def test2():
    x=100#local 
    print("inside test2------>x=",x)

def adder():
    global x
    x+=1
    print("inside adder++++++x=",x)
print("outside :x=",x)
test()
test2()
adder()
print("outside :x=",x)
```
