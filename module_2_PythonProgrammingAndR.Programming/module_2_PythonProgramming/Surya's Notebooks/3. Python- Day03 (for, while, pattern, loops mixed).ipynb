```python
#         5
#       4 5
#     3 4 5
#   2 3 4 5
# 1 2 3 4 5
space=5
i=1
while i<=5:
  s=1
  while s<=space: #space printing
    print(" ", end=" ")
    s+=1
  j=1
  while j<=i:
    print(j, "", end="")
    j+=1
  i+=1
  space-=1
  print()
```

```python
# interview challange capgemini, to print armstrong numbers between 100-999
# 153, 370, 371, 407
for no in range(100, 1000):
  tno=no
  sum=0
  while tno>0:
    d=tno%10
    tno=tno//10
    sum=sum + d**3
  if(sum==no):
    print(no)
```

```python
# interview challange, TCS, to check if given number is part of fibonacci series or not
n=int(input("Enter a number you want to check in fibonacci series: "))
flag=0
f0,f1=0,1
for i in range(0, n+1):
  if i<=1:
    if i==n:
      print(n, "is a part of fibonacci series")
      flag=1
      break
  else:
    fn=f0+f1
    f0, f1=f1, fn
    if fn==n:
       print(n, "is a part of fibonacci series")
       flag=1
    if fn>n:
      break
if flag==0:
  print(n, "is not a part of fibonacci series")
```

```python
# print prime numbers between 1-100
for no in range(1, 101):
  i, flag=2, False
  while i<no/2:
    if(no%i==0):
      flag=True
      break
    i+=1
  if flag==False:
    print(no)
```
