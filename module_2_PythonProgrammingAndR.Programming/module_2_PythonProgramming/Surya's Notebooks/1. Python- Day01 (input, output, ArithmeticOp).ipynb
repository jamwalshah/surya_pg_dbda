Python supports dynamic casting, it means data type of a variable will be decided by the data given to it

```python
# swapping without 3rd variable
a=10
b=20
print("a: ", a, "b: ", b)
a, b=b, a
print("a: ", a, "b: ", b)
```

```python
a=10
b=5*2
c=13-3
print("a:", id(a), "b: ", id(b), "c: ", id(c))
a=a-5
print("a:", id(a), "b: ", id(b), "c: ", id(c))
b=b/2
print("a:", id(a), "b: ", id(b), "c: ", id(c))
```

```python
a, b, c=10, 20, 30
print(a, b, c, sep="------")
print(a, b, c, sep="------")
print(a, b, c, sep="------")
```

```python
a, b, c=10, 20, 30
print(a, b, c, sep="------", end=" ") # end with single space, & separator as "------"
print(a, b, c, sep="------")
print(a, b, c, sep="------")
```

```python
a, b, c=10, 20, 30
print(a, b, c, sep="------", end="\t") # end with tab space (5 spaces)
print(a, b, c, sep="------")
print(a, b, c, sep="------")
```

```python
no =input("Enter a number:") # treats input as 'str'
print(no, type(no))
```

```python
no =int(input("Enter a number:")) # casts 'str' input to 'int'
print(no, type(no))
```

```python
# challange
# read boolean False from keyboard
no =bool(input("Enter a number:")) # casts 'str' input to 'bool'
print(no, type(no))   # Abscence of information is False, so no input will read False
```
