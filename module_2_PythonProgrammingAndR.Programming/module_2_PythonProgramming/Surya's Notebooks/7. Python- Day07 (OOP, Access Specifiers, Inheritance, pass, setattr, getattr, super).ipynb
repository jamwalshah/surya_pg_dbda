```python
# CG, 3i, and morningstar
# id, name, salary --strong private
# setDetail() , printDetail() --private
# change setDetails to constructor
# replace printDeails with printer
# empID should be auto generating
# store details of 5 employees
# menu driven to create, search, delete employee
class Emp:
    empid=202300
    def __init__(self, name, salary):
        self.__id=Emp.empid
        print("Employee ID: ", self.__id, "\n")
        self.__name=name
        self.__salary=salary
        Emp.empid+=1
    def __str__(self):
        return "ID: "+str(self.__id)+" Name: "+self.__name+" Salary: "+str(self.__salary)
    def getempid(self):
        return self.__id

emplist=[]
while True:
    print("1. Create \n2. Search\n3. Delete\n0. Exit")
    ch=int(input(" : "))
    if ch ==1:
        e=Emp(input("\nEnter Employee Name: "), float(input("Enter Salary: ")))
        emplist.append(e)
    elif ch==2: # search logic
        tempid=int(input("Enter ID: "))
        flag=False
        for employee in emplist:
            if employee.getempid()==tempid:
                print("Found")
                flag=True
                print(employee)
                break
        if flag==False:
            print("Not found")
    elif ch==3: # delete logic
        # read id to delete
        tempid=int(input("Enter ID: "))
        flag=False
        for clsele in range(len(emplist)):
            if emplist[clsele].getempid()==tempid:
                print("Found")
                flag=True
                print("Deleted: ", emplist.pop(clsele))
                break
        if flag==False:
            print("Not found")
    elif ch==0:
        print("Bye-Bye")
        break
    else:
        print("\nInvalid input\n")
```

##super

```python
# super
class A:
  def __init__(self):
    print("A")
class B(A):
  def __init__(self):
    super().__init__() # will call __init__() from super/parent class
    print("B")
obj=B()
```

```python
# super in case of inheritance
class A:
  def __init__(self):
    print("A")
class B:
  def __init__(self):
    print("B")
class X(A, B):
  def __init__(self):
    super().__init__() # will call __init__() class A, by following FCFS in inheritance
    print("X")
obj=X()
```

```python
# super
class A:
  def __init__(self):
    print("A")
class B:
  def __init__(self):
    print("B")
class X(B, A):
  def __init__(self):
    super().__init__() # will call __init__() class B, by following FCFS in inheritance
    print("X")
obj=X()
```

```python
# super -  calling parent method from child class
class A:
  def __init__(self, d1):
    print("A", d1)
class B(A):
  def __init__(self, d1, d2):
    super().__init__(d1) # will call __init__() class A, by following FCFS in inheritance
    print("B", d2)
class X(B):
  def __init__(self, d1, d2, d3):
    super().__init__(d1, d2) # will call __init__() class B, by following FCFS in inheritance
    print("X", d3)
obj=X(11, 22, 33)
```

```python
#create a bank class 
# has amount,accountno,name
# createaccount(constructor)-user only gives name and amount 
# account number auto generated
#withdraw(amount):should not be -ve amount and min 
# balance is 2000 else reject transection
#deposite(amount):amount can not be -ve
#checkbalance():shows balance and account holder name

'''
menu driven code to 
1 create account
2 withdraw
3 deposite
4 check balance

1---->create by takingvalues and auto generate account number
2/3/4--->ask account number ,search account and then operate
'''
# author: Surya Dev Singh Jamwal
# date: 06 Apr 2023
# Title/purpose: mini bank-teller app
def notnegamtchk(amt):
    return amt>=0

class Bankacc:
    sban=4000
    def __init__(self, acholdernam, idprfnum):
        Bankacc.sban+=1
        self.__sbaccn=Bankacc.sban
        print("New Account created, SB Acc No :", self.__sbaccn)
        self.__accholdername=acholdernam
        self.__idproofnumber=idprfnum
        self.__balance=2500
        print("Rs. 2500 credited upon account opening of a/c no.", self.__sbaccn)
    
    def getaccno(self):
        return self.__sbaccn

    def withdrawalction(self, wamt, accno):
        if notnegamtchk(wamt):
            if self.__balance-wamt>2000:
                self.__balance-=wamt
                print("Withdrawal of", wamt, "from a/c", self.__sbaccn, ", updated balance:", self.__balance)
            else:
                print("A/c no:", self.__sbaccn, "has Insufficient Balance")
    
    def depositaction(self, damt, accno):
        if notnegamtchk(damt):
            self.__balance+=damt
            print("Deposit of", damt, "from a/c", self.__sbaccn, ", updated balance:", self.__balance)

    def balchk(self):
        print("a/c no:", self.__sbaccn, "balance:", self.__balance)
    

sbal=[] # list to store accounts
def validateaccnum(an):
    for ele in sbal:
        if ele.getaccno()==an:
            return True
            break
    return False

while True:
    print("\n0. Exit\n1. Create New Account\n2. Withdrawal\n3. Deposit\n4. Check Balance")
    inp=int(input(" : "))
    if inp==1: # create new account
        ba=Bankacc(input("Enter account holder name:"), int(input("Enter ID proof number:")))
        sbal.append(ba)
    elif inp==2: # Withdrawal
        an=int(input("Enter account number to withdraw: "))
        if validateaccnum(an):
            amt=int(input("Enter withdrawal amount : "))
            for ele in range(len(sbal)):
                if sbal[ele].getaccno()==an:
                    sbal[ele].withdrawalction(amt, an)
        else:
            print("Invalid account number")
    elif inp==3: # deposit
        an=int(input("Enter account number to deposit: "))
        if validateaccnum(an):
            amt=int(input("Enter deposit amount : "))
            for ele in range(len(sbal)):
                if sbal[ele].getaccno()==an:
                    sbal[ele].depositaction(amt, an)
        else:
            print("Invalid account number")
    elif inp==4: # check balance
        an=int(input("Enter account number to Check balance: "))
        if validateaccnum(an):
            for ele in range(len(sbal)):
                if sbal[ele].getaccno()==an:
                    sbal[ele].balchk()
        else:
            print("Invalid account number")
    elif inp==0:
        print("\nExiting Program, Byee")
        break
    else:
        print("\nInvalid Input, Try Again")
```
