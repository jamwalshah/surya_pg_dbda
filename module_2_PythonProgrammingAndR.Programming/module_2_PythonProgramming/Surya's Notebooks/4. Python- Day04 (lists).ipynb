```python
# enter all elements in a list first blank
# then print sum of all , use sum()
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print(list1)
print("sum: ", sum(list1))
```

```python
# enter all elements in a list first blank
# print all elements lesser than the average 
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print(list1)
average=sum(list1)/len(list1)
print("Average: ", sum(list1)/len(list1))
for elem in list1:
  if elem < average:
    print(elem, end=" ")
```

```python
# enter all elements in a list first blank
# print all elements lesser than the average in sorted order
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print(list1)
average=sum(list1)/len(list1)
print("Average: ", sum(list1)/len(list1))
list1.sort()
for elem in list1:
  if elem < average:
    print(elem, end=" ")
```

```python
# enter all elements in a list first blank
# print second largest & second smallest
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print(list1)
list1.sort()
print(list1)
print("second largest:", list1[-2])
print("second smallest:", list1[1])
```

```python
# [1, 2, 3, 4, 5]
#rotate content for n time given by user
# pass1: [2, 3, 4, 5, 1]
# pass2: [3, 4, 5, 1, 2]

list1=[1, 2, 3, 4, 5]
print(list1)
i=int(input("Enter clock cycles: "))
while i>0:
  n=list1.pop(0)
  list1.append(n)
  print(list1)
  i-=1
```

```python
# [1, 2, 3, 4, 5]
#rotate content anti-clockwise for n time given by user
# pass1: [2, 3, 4, 5, 1]
# pass2: [3, 4, 5, 1, 2]
list1=[1, 2, 3, 4, 5]
print(list1)
i=int(input("Enter clock cycles: "))
while i>0:
  n=list1.pop(4)
  list1.insert(0, n)
  print(list1)
  i-=1
```
