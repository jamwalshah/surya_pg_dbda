```python
class Emp:
    count=202300
    def __init__(self,name,salary):
        Emp.count+=1
        self.__id=Emp.count
        self.__name=name
        self.__salary=salary
        print("Record added:ID-->",self.__id)
    def __str__(self):
        return str(self.__id)+" "+self.__name+" "+str(self.__salary)
    def getempid(self):
        return self.__id
```

```python
emplist=[]
while True:
    print("1.Create\n2.Search\n3.Delete\n0.Exit")
    ch=int(input(":"))
    if ch==1:
        n=input("Enter name:")
        s=float(input("salary:"))
        e=Emp(n,s)
        emplist.append(e)
    elif ch==2:
        tempid=int(input("Enter id:"))
        flag=False
        for i in emplist:
            if i.getempid()==tempid:
                print("Found:")
                flag=True
                print(i)
                break
        if flag==False:
            print("Not Found")
    elif ch==3:
        tempid=int(input("Enter id:"))
        flag=False
        for i in range(len(emplist)):
            if emplist[i].getempid()==tempid:
                print("Found:")
                flag=True
                print("Deleted:",emplist.pop(i))
                break
        if flag==False:
            print("Not Found")

    elif ch==0:
        print("bye")
        break
    else:
        print("Wrong choice given")
    
    
            
```

```python
#super()---->allow access of super class's methods
class A:
    def __init__(self,d1):
        print("A",d1)
class B(A):
    def __init__(self,d1,d2):
        super().__init__(d1)
        print("B",d2)
class X(B):
    def __init__(self,d1,d2,d3):
        super().__init__(d1,d2)
        print("X",d3)
obj=X(11,22,33)
```

```python
class triangle:
    def setLH(self,l,h):
        self.l=l
        self.h=h
    def area(self):
        print("Area is:",(0.5*self.l*self.h))
```
