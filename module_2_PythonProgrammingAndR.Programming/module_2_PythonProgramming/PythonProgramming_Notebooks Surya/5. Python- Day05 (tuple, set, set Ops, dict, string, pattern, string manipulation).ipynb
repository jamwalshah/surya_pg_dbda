```python
# enter all elements in a list first blank
# then print sum of all , use sum()
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print("unique data\n")
for i in sorted(set(list1)):
  print(i)
```

```python
# list all elements in sorted manner with their count
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print("unique data\n")
for i in sorted(set(list1)):
  print(i, "frequency: ", list1.count(i))
```

```python
# list all elements in sorted manner with their count, also print the one with highest count
list1=[]
while True:
  ele=input("Enter element: ")
  if ele=="":
    break
  list1.append(float(ele))
print("unique data\n")
hig=0
higel=0
for i in sorted(set(list1)):
  print(i, "frequency: ", list1.count(i))
  if list1.count(i)>hig:
    hig=list1.count(i)
    higel=i
print("Highest frequency:", hig, "element:", higel)
```

```python
# print number in words, digit by digit
# e.g. 123 --> one two three
n=int(input("Enter a number: "))
di={0:"zero", 1:"one", 2:"two", 3:"three", 4:"four", 5:"five", 6:"six", 7:"seven", 8:"eight", 9:"nine"}
no=n
rev=0
while no>0:
  rev=(rev*10)+(no%10)
  no=no//10
while rev>0:
  d=rev%10
  rev=rev//10
  print(di[d], end=" ")
```

```python
# take a number and print unique digits only in sorted manner in words
# input: 361421288
# output: one two three four six eight
n=int(input("Enter a number: "))
di={0:"zero", 1:"one", 2:"two", 3:"three", 4:"four", 5:"five", 6:"six", 7:"seven", 8:"eight", 9:"nine"}
se=set()
no=n
while no>0:
  d=no%10
  rev=(rev*10)+(no%10)
  no=no//10
  se.add(d)
for i in sorted(se):
  print(di[i], end=" ")
```

```python
# print key for max marks
d={"m1":77, "m2":67, "m3":57, "m4":97, "m5":7, "m6":47}
kys=d.keys()
max=0
msub=''
min=9999
mnsub=''
for i in kys:
  if d[i]>max:
    max=d[i]
    msub=i
  if d[i]<min:
    min=d[i]
    mnsub=i
print("Maximum marks:", max, "in subject:", msub)
print("Minimum marks:", min, "in subject:", mnsub)
```
