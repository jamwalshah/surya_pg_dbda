```python
# if no1%no2==0 then no2 is gcd, else let no1=no2 and no2 be no1%no2
def gcd(no1, no2):
  if no1%no2==0:
    return no2
  else:
    return gcd(no2, no1%no2)
```

```python
print(gcd(12, 18))
```

```python
def Descending(no1, no2, no3):
  ''' takes 2 inputs and prints maximum of 2'''
  if no1>no2 and no1>no3:
    if no2>no3:
      return no1, no2, no3
    else:
      return no1, no3, no2
  elif no2>no1 and no2>no3:
    if no1>no3:
      return no2, no1, no3
    else:
      return no2, no3, no1
  elif no3>no1 and no3>no2:
    if no1>no2:
      return no3, no1, no2
    else:
      return no3, no2, no1
      
```

##generator

```python
def fd(amount, roi): # generator
  y=1
  print("at end of a year")
  yield(amount*(roi/100)*y)
  y+=1
  print("at end of 2nd year")
  yield(amount*(roi/100)*y)
  y+=1
  print("at end of 3rd year")
  yield(amount*(roi/100)*y)
  y+=1
```

```python
i=fd(1000, 7)
```

```python
next(i)
```

##variable scope

```python
# global variable accessed by all
x=10
def test():
  print("inside test x: ", x)
def test2():
  x=100
  print("inside test2 ---> x:", x)

print("outside: x:", x)
test()
test2()
```

```python
# global variable accessed by all
x=10
def test():
  print("inside test x: ", x)
def test2():
  x=100
  print("inside test2 ---> x:", x)
def adder():
  global x
  x+=1
  print("inside adder +++++++x:", x)
print("outside: x:", x)
test()
test2()
adder()
```

```python
class Human:
  def birth(self):
    print("Reference:", id(self))
    self.gender=input("Enter gender: ")
  def naming(self):
    self.name=input("Enter name: ")
  def intro(self):
    print("Hi I am a", self.gender, "called", self.name)
```

```python
class Human:
  def __init__(self): # constructor
    print("object created:", id(self))
    self.gender=input("Enter gender: ")
    self.name=input("Enter name: ")
  def intro(self):
    print("Hi I am a", self.gender, "called", self.name)
```

```python
h=Human()
```

```python
h.intro()
```

```python
print(h)
```

```python
class Human:
  def __init__(self): # constructor method
    print("object created:", id(self))
    self.gender=input("Enter gender: ")
    self.name=input("Enter name: ")
  def __str__(self): # printer method, returns string
    return "Hi I am a "+self.gender+" called "+self.name
```

```python
h=Human()
```

```python
print(h)
```
