##Exception Handling

```python
#we are only trapping error, not what type of error
a=int(input("Enter a number: "))
b=int(input("Enter another number: "))
c=a/b
print("C is", c)
print("Code is running")
print("Coded by amarpanchal.education")
```

```python
#we are only trapping error, not what type of error
a=int(input("Enter a number: "))
b=int(input("Enter another number: "))
try:
    c=a/b 
except Exception:
    print("Exception")
else:
    print("C is", c)
finally:
    print("by Amar sir")
    # part of code after handling exception
print("still code is running")
```

```python
#we are trapping error as well as the type of error
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b 
except Exception as arg:
    print("Exception:", arg)
else:
    print("C is", c)
finally:
    print("by Amar sir")
    # part of code after handling exception
print("still code is running")
```

```python
#executing with only 'try' & 'finally' blocks which are mandatory
# creates a risk that it cannot catch error
try:
    a=int(input("Enter a number: "))
    b=int(input("Enter another number: "))
    c=a/b 
    print("C is", c)
finally:
    print("finally executed")
    # part of code after handling exception
print("still trying to live life")
print("alive and kicking")
```

```python
# create custom exception
class AgeException(Exception): # inherit super class Exception to create custom exception
    def __init__(self, data):   # accept the exception which is raised
        self.data=data 
    def __str__(self):  # throws the exception
        return "Exception : "+str(self.data)+" Under 18"

try:
    n=int(input("Enter age: "))
    if n>=18:
        print("You can vote, your age is", n)
    else:
        raise AgeException(n)   # raises exception
except Exception as msg:    # excepting exception to throw it
    print(msg)
finally:
    print("code is executing at last step, now stopping")
```
