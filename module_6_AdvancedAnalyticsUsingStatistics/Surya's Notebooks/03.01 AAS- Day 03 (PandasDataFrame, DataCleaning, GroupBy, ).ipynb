````python
import numpy as np
import pandas as pd
# importing NumPy & pandas
````

### DataFrame.groupby('ColumnName')
- groupby: Grouped information based on columns
- groups created can be used to perfrom aggregate functions

````python
grp1 = df.groupby('Location')
# creates a DataFrame GroupBy on Location column
grp1
# prints object information for DataFrame GroupBy
````

### DataFrameGroupBy.groups

````python
grp1.groups
# gruoups property lsts down all the groups in a DataFrame Group
````

````python
len(grp1.groups)
# prints count of unique groups in the column where GroupBy is created
````

````python
type(grp1)
# type of groupby object is DataFrameGroupBy
````

### DataFrameGroupBy.get_group('UniqueValue')

````python
grp1.get_group('MWH-2')
# uses DataFrameGroupBy.get_group property to fetch records
# for a specified unique value / group only
````

### DataFrameGroupBy.agg(AggregateFunction)
- works with DataFrameGroupBy, but returns a DataFrame

````python
grp1.agg(np.mean)
# can specify the aggregate function to be used with DataFrame GroupBy
# printing the mean of each of the groups created on location column
````

````python
grp1.agg(np.size)
# printing the size/length of each of the groups created on location column

# NOTE that, now it prints size for MaterialID column too,
# because it can count all type of values,
# but for numeric aggregate functions like mean, etc. ,
# it takes up column which have numeric values only
````

````python
grp1.agg(np.sum)
# printing the sum of each of the groups created on location column

# Here again, it'll take coulumns which have numeric values only
# for numeric functions like sum, mean
````

### DataFrameGroupBy.agg([ListOfAggreagateFunctions])

````python
grp1.agg([np.sum, np.mean])
# printing the sum & mean of each of the groups created on location column
# but it'll work on numeric columns only
````

````python
r = grp1.agg([np.sum, np.mean])
# returns a DatFrame from DaaFrameGroupBy using aggregate functions
r
# printing DataFrame r
````

````python
type(r)
# checking type of DataFrame containing aggregate information
#  returned by DataFrameGroupBy.agg()
````

````python
ind = np.where(r.index == 'MWH-3')
# using np.where() to find the indices of records of
# DataFrame which has index = 'MWH-3'
ind
# printing indices of records satisfying the condition
````

````python
r.iloc[ind]
# using iloc on DataFrame to print the records on indices
# satisfying the condition
````

````python
grp2 = df.groupby(['MaterialID', 'Location'])
# creating another group using DataFrame.groupby() to make
# groups by 'MaterialID' column, and then sub-grouping by 'Location' column
grp2.groups
````

````python
grp3 = df.groupby(['Location', 'MaterialID'])
# creating another group using DataFrame.groupby() to make
# groups by 'Location' column, and then sub-grouping by 'MaterialID' column
grp3.groups
````

````python
grp2.agg([np.sum, np.size])
# using aggregate functions on DataFrameGroupBy which is grouped
# on two columns of DataFrame,
# so there will be sub-groups for groups,
# and then it'll execute aggregate functions for each of the sub-groups

# Note that, here the order of grouping is 'MaterialID' then 'Location'
# groups are arranged aplhabetically, so access time is reduced
# and DataFrame becomes optimized
````

````python
grp3.agg([np.sum, np.mean])
# Note that, here the order of grouping is 'Location' then 'MaterialID'
````

### pd.merge with 'Pune' & 'Mumbai'

````python
import numpy as np
import pandas as pd
````

### pd.read_excel('filename.xlsx', sheet_name='WorkSheet')

````python
df1 = pd.read_excel('locations.xlsx', sheet_name='Pune')
# taking workSheet 'Pune' from WorkBook 'locations.xlsx' into DataFrame
df2 = pd.read_excel('locations.xlsx', sheet_name='Mumbai')
# taking workSheet 'Mumbai' from WorkBook 'locations.xlsx' into DataFrame
````

````python
df1
# printing DataFrame df1
````

````python
df2
# printing DataFrame df2
````

### pd.merge(DataFrame1, DataFrame2, on='CommonColumnName')
- produces inner join of DataFrame1 & DataFrame2 on common column

````python
pd.merge(df1, df2, on='Subject')
# returns resultant DataFrame which joins df1 with df2 on column subject
# produces inner join by default
````

### pd.merge(DataFrame1, DataFrame2, on='CommonColumnName', how='JoinType')

````python
pd.merge(df1, df2, on='Subject', how='inner')
# returns resultant DataFrame which joins df1 with df2
# on column 'subject' by 'inner join'

# performs intersection of two DataFrames
````

````python
pd.merge(df1, df2, on='Subject', how='outer')
# returns resultant DataFrame which joins df1 with df2
# on column 'subject' by 'outer join'

# performs union of two DataFrames
````

````python
pd.merge(df1, df2, on='Subject', how='left')
# returns resultant DataFrame which joins df1 with df2
# on column 'subject' by 'left join'
````

````python
pd.merge(df1, df2, on='Subject', how='right')
# returns resultant DataFrame which maps df1 with df2
# on column 'subject' by 'right join'
````

````python
df1
# printing actual DataFrame df1
````

````python
df2
# printing actual DataFrame df2
````

### pd.merge(DataFrame1, DataFrame2, on=[listOfColumns], how='JoinType')

````python
pd.merge(df1, df2, on=['Subject', 'Grade'], how='inner')
# returns resultant DataFrame which joins df1 with df2
# on two columns 'Subject' & 'Grade' by 'inner join'
````

````python
pd.merge(df1, df2, on=['Subject', 'Grade'], how='outer')
# returns resultant DataFrame which joins df1 with df2
# on two columns 'Subject' & 'Grade' by 'outer join'
````

````python
pd.merge(df1, df2, on=['Subject', 'Grade'], how='left')
# returns resultant DataFrame which joins df1 with df2
# on two columns 'Subject' & 'Grade' by 'left join'
````

````python
pd.merge(df1, df2, on=['Subject', 'Grade'], how='right')
# returns resultant DataFrame which joins df1 with df2
# on two columns 'Subject' & 'Grade' by 'right join'
````
