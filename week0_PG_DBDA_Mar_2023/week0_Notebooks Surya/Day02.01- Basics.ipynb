##Rules for identifier
1. we cannot start an identifier name with integer or digits.
e.g. 1var = 10 # is invalid
2. we cannot use special symbols
e.g. var@12 = 34 # is invalid
3. '_' is allowed
e.g. var_12 = 34 # is valid
4. keywords are not allowed as identifier
e.g. import = 123 # is invalid
5. alpha-numeric keyword is allowed
