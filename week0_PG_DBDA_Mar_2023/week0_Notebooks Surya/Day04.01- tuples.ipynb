##tuple
1. tuple is similar to list
2. tuples are immutable, cannot be changed
3. it is in ordered format, like list
4. iteration is faster in tuples as compared to list

```python
t = () #empty tuple
```

```python
type(t)
```

```python
t = (1, 2, 3, 4) #tuple of int
t
```

```python
t = (1.1, 21, 3.100, 4.100) #tuple of float
t
```

```python
t = ("one", "two", "four", "five") #tuple of string
t
```

```python
t = ("one", "two", 12, 34, 2.4, 4.5) # mix data types tuples
t
```

```python
t = ("one", "two", (12, 34), 2.4, 4.5, [1, 3, 4], {6, 8, 23, 45})
t
```
