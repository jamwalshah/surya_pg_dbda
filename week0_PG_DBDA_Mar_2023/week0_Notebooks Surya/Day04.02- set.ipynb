2. sets does not allow duplicate elements
3. set is im-mutable, cannot be changed

set does not allow any mutable type of element such as list, etc. to be nested in a set, because set is im-mutable

```python
s = {"three", "two", "one", 1.2, 3.1, 5.4, 1, 2, 3}
s
```

```python
s.add("four") #adds only single element
s
```

```python
s.update(["six", 1, 3, 45]) #adds multiple elements
s
```

```python
s.remove("four")
s
```

```python
s.discard("two")
s
```

Set operation

1. Union Op set1|set2 : removes duplicates and returns the combined set

```python
s = {1, 3, 4, 67, 8}
s1 = {21, 4, 3, 687, 798, 4}
s2 = {567, 34, 67, 23}
s3 = {57, 4, 7, 3}
```

```python
s|s1 #union op : removes duplicates and returns the combined set
```

```python
s.union(s1)
```

```python
s.union(s1, s2, s3)
```

2. intersection Op set1&set2: common elements only

```python
s = {1, 3, 4, 67, 8}
s1 = {21, 4, 3, 687, 798, 4}
s2 = {567, 34, 67, 23, 4}
s3 = {57, 4, 7, 3}
```

```python
s & s1
```

```python
s & s1 & s2 &s3
```
