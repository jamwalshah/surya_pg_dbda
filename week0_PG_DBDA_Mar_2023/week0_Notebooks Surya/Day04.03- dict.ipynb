```python
keys = {1, 2, 3, 4}
dic_new = dict.fromkeys(keys) # creats dict with only keys and None values
dic_new
```

```python
keys = {1, 2, 3, 4}
values = 90
dic_new = dict.fromkeys(keys,values) # creates dict with keys and value as 90 for all
dic_new
```

```python
keys = {1, 2, 3, 4}
values = [10, 20, 30]
dic_new = dict.fromkeys(keys,values) # creates dict with keys and value as [10, 20, 30] for all
dic_new
```

```python
values.append(12)
dic_new = dict.fromkeys(keys,values) # adds value 12 in each of the list in value set
dic_new
```

```python
#dic_new[0] # throws error if key does not exist
```

```python
dic_new[3] # prints value for the key passed as index, key is treated as index in dict
```

```python
dic_new.get(4) # returns the value for key passed to get()
```

```python
dic = {"A":"one", "B":"two", "C":"three"}
dic
```

```python
dic["A"] = "ONE" # replaces value for char-type key specified as index
```

```python
dic
```

```python
dic_1 = {"B":"Second"}
dic.update(dic_1) # use another key-value pair as an argument to update() method to update/create value for a key
dic
```

```python
dic_1 = {"D":"Four"}
dic.update(dic_1) # use another key-value pair as an argument to update() method to update/create value for a key
dic
```

.update(var) will update the value for the key-value pair as specified, and if that key does not exist, it will add a new key-value pair as specified
