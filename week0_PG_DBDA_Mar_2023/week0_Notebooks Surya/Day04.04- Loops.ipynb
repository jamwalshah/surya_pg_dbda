## Loops
for: condition in form of (init, increment, limit/exit condition)

```python
n = [1, 2, 3, 4, 5]
for i in n:
  print(i) # prints all elements from list
```

```python
for i in range(5): # range(end) , by default starts from zero
  print(i)
```

```python
for i in range(1, 5):
  print(i) # prints numbers 1 to 4 as range starts from 1
```

```python
n = int(input("Enter  number : "))
for i in range(1, n+1): # range(start, end)
  print(i) # prints numbers till user has specified 
```

```python
n = int(input("Enter  number : "))
for i in range(2, n+1, 2): # range(start, end, steps) # zero is neither even nor odd
  print(i)  # prints even numbers
```

```python
n = int(input("Enter  number : "))
for i in range(1, n+1, 2): # range(start, end, steps)
  print(i)  # prints odd numbers
```

```python
n = input("Enter your name : ")
for i in range(len(n)):
  print(n[i]) # separates chars from string
```

```python
n = input("Enter string : ")
for i in range(2, len(n)):
  if i%2 == 0:
    print(n[i]) # prints chars at even indices
```

```python
n = input("Enter string : ")
for i in range(2, len(n), 2):
  print(n[i]) # prints chars at even indices
```

```python
num = [20, 34, 65, 12, 67, 89]
for i in range(len(num)):
  if num[i] % 2 == 0 :
    print(num[i]) # prints even elements from list
```

```python
num = [20, 34, 65, 12, 67, 89]
for i in range(len(num)):
  if num[i] % 2 == 0 :
    print("Even number", num[i]) # prints even elements from list
  else:
    print("Odd Number", num[i]) # prints odd elements from list
```
