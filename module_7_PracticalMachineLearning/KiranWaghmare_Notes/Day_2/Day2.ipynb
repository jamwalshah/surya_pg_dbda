```python
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('Graph in 2-Dimension')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed')
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('Graph in 2-Dimension')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*--')

plt.subplot(2,2,3)
plt.plot(x,y,'r--')

plt.subplot(2,2,4)
plt.plot(x,y,'r--')
```
