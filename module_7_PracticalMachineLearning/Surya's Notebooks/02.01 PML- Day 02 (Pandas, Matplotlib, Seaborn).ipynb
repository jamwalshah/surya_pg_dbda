# matplotlib library

## import libs

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## create X & Y

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
# creating ndarrays
```

### scatter plots

#### plt.scatter(x, y)

```python
plt.scatter(x, y)
# plot a scatter graph
```

#### plt.scatter(x, y, c='g')

```python
plt.scatter(x, y, c='g')
# plot a scatter graph with color
```

```python
plt.scatter(x, y, c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2-D')
# plot a scatter graph with color xlabel, ylabel & title
```

#### plt.plot(x, y, format, linestyle)

```python
# x=np.arange(1000)
y=y**2
plt.plot(x, y, 'r*', linestyle='dashed')
```

#### plt.subplot(int, int, index)

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'g*--')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'g--')
```

```python
y=3*x + 5
plt.plot(x, y, 'y*--')
```

```python
np.pi
```

```python
x = np.arange(1, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, 'r*--')
# plots sin-wave
```

```python
x = np.arange(1, 4*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y, 'r*--')
# plots cos-wave
```

# seaborn library

## import libs

```python
import seaborn as sns
```

## import dataset from seaborn

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

#### relations / correlations

##### df.corr()

```python
df.iloc[ : , [0, 1, 6]].corr()
# correlation matrix
```

##### heatmap(correlation)

```python
sns.heatmap(df.iloc[ : , [0, 1, 6]].corr())
# heatmap
```

##### pairplot

```python
sns.pairplot(df)
```

##### jointplot

```python
sns.jointplot(x='tip', y='total_bill', data=df)
# sns.jointplot(x, y, data=df)
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
# sns.jointplot(x, y, data=df, kind='hex')
```

```python
sns.pairplot(df, hue='sex')
```

### distribution plots

#### distplot

```python
# distplot
sns.distplot(df['tip'])
```

#### displot

```python
# displot
sns.displot(df['tip'])
```

### Categorical plots

#### countplot()

```python
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

#### boxplot()

```python
# box plot
sns.boxplot(x='smoker', y='total_bill', data=df)
# sns.boxplot(x='a', y='b', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
# sns.boxplot(x='c', y='b', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette ='rainbow')
# sns.boxplot(x='a', y='b', data=df, palette ='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
# sns.boxplot(data=df, orient='v')
```

#### violinplot()

```python
sns.violinplot(x='day', y='total_bill', data=df, palette='rainbow')
# sns.violinplot(x='a', y='b', data=df, palette='rainbow')
```

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
# sns.violinplot(x='b', y='a', data=df, palette='rainbow')
```
