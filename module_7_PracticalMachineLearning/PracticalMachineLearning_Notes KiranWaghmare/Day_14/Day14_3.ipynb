# DBSCAN Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.DataFrame(np.load('D:\Test\D14data2.npy'),columns=['x', 'y'])
dataset
```

```python
dataset.plot(kind='scatter', x='x', y='y')
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.033,
               min_samples=22,
               algorithm='brute',
               p=0.9)

```

```python
db = dbscan.fit_predict(dataset)
```

```python
la = dbscan.labels_
la
```

```python
len(dbscan.core_sample_indices_)
```

```python
dataset.plot(kind='scatter', x='x', y='y', c=db,cmap='Accent_r', figsize=(16,10))
plt.show()
```
