# Clustering Algorithm : DBSCAN Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('D:/Test/D14data1.csv')
dataset
```

```python
#Input dataset
X = dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
X
```

```python
#Model Building
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3,min_samples=4, metric='euclidean')
cls = dbscan.fit(X)

```

```python
la = cls.labels_
la
```

```python
y_pred = dbscan.fit_predict(X)
y_pred
```

```python
plt.scatter(X[y_pred == 0, 0],X[y_pred == 0, 1], s=50, c='red',label = 'C1')
plt.scatter(X[y_pred == 1, 0],X[y_pred == 1, 1], s=50, c='blue',label = 'C2')
plt.scatter(X[y_pred == 2, 0],X[y_pred == 2, 1], s=50, c='green',label = 'C3')
plt.scatter(X[y_pred == 3, 0],X[y_pred == 3, 1], s=50, c='yellow',label = 'C4')
plt.scatter(X[y_pred == 4, 0],X[y_pred == 4, 1], s=50, c='cyan',label = 'C5')
plt.scatter(X[y_pred == 5, 0],X[y_pred == 5, 1], s=50, c='gray',label = 'C6')
plt.scatter(X[y_pred == 6, 0],X[y_pred == 6, 1], s=50, c='pink',label = 'C7')
plt.scatter(X[y_pred == 7, 0],X[y_pred == 7, 1], s=50, c='magenta',label = 'C8')
plt.scatter(X[y_pred == 8, 0],X[y_pred == 8, 1], s=50, c='orange',label = 'C9')
plt.title('KMeans Clustering ')
plt.xlabel('Customer score')
plt.ylabel('Customer Income')
plt.legend()
plt.show()
```
